```python
#Importação das bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Importação de dados
df = pd.read_csv('wine_dataset.csv')
df.head()
```

# PROCESSAMENTO DE DADOS

```python
#Convertendo tipo de dados objeto para inteiro
df['style'] = df['style'].replace('red',0)
df['style'] = df['style'].replace('white',1)
df.head()
```

```python
#Dividindo features(X) e target(y)
X = df.drop('style',axis=1)
y = df['style']
```

```python
#Separação de dados de treino e teste 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

# CONSTRUÇÃO MÁQUINA PREDITIVA

```python
#Importando biblioteca e construindo máquina preditiva
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(X_train,y_train)
result = modelo.score(X_test,y_test)
print('Precisão:',result )
```

```python
#Análise de Máquina preditiva
print(y_test[0:5])
previsao = modelo.predict(X_test[0:5])
print(previsao)
```
